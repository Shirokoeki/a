```python
prom =[]
n = 10
cont = 0
sum = 0

for i in range(n):
  num = int(input("Numero: "))
  sum += num
  prom.insert(cont, num)
  cont += 1
prome = sum / 10
print("Esta es la suma: ",sum)
print("Este es el promedio: ",prome)
```
